```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([[0,0,1,1], [0,1,0,1]])
# XOR:
# y = np.array([[0,1,1,0]])
# XNOR:
y = np.array([[1,0,0,1]])
no_x = 2
no_y = 1
no_h = 2
tot = x.shape[1]
lr = 0.1
np.random.seed(2)

w1 = np.random.rand(no_h, no_x)
w2 = np.random.rand(no_y, no_h)
losses = []

def sigmoid(z):
    z = 1/(1+np.exp(-z))
    return z

def frwd_prop(w1, w2, x):
    z1 = np.dot(w1, x)
    a1 = sigmoid(z1)
    z2 = np.dot(w2, a1)
    a2 = sigmoid(z2)
    return z1, a1, z2, a2

epochs = 20000

for i in range(epochs):
    z1, a1, z2, a2 = frwd_prop(w1, w2, x)
    loss = -(1/tot)*np.sum(y*np.log(a2)+(1-y)*np.log(1-a2))
    losses.append(loss)

plt.plot(losses)
plt.xlabel("Epochs")
plt.ylabel("Loss Value")
```
